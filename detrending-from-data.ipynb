# Detrending

```python
import numpy as np
import yfinance as yf
tesla = yf.download("TSLA")
price = tesla["Adj Close"]
```

```python
price.plot()
```

```python
trend = price.rolling(window=14).mean()
price.rolling(window=14).mean().dropna().plot()
```

```python
detrend = price - trend
detrend.plot()
```

```python
import matplotlib.pyplot as plt
plt.plot(detrend[-1000:])
```

```python
tesla = yf.download("TSLA")
price = tesla["Adj Close"]
```

```python
np.polyfit(range(len(price[price.index>='2018-01-01'])), price[price.index>='2018-01-01'], deg=3)
```

```python
y = np.polyfit(range(len(price)), price, deg=16) 
z = np.polyval(y, range(len(price)))
```

```python
plt.plot(price.values)
plt.plot(z)
```

```python
trend = price- z
trend.plot()
trend
```
